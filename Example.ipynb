```python
import pandas as pd
import chorogrid
```

# State by State Data

```python
state_data = pd.read_csv('chorogrid/sample_data/sample_state_data.csv')
```

```python
for shape in ("hex", "multihex", "square", "map"):
    # equivalent to:
    # chorogrid.plot(state_data.state, state_data.Percent_living_in_same_home_as_one_year_ago, ...)
    state_data.plot.choropleth(
        x="state", y="Percent_living_in_same_home_as_one_year_ago",
        title="Percent Living in Same Home", legend="percent",
        shape=shape)
```

# County Level Data

```python
county_data = pd.read_csv('chorogrid/sample_data/sample_county_data.csv', encoding='latin-1')
```

```python
county_data.plot.choropleth(
    # equivalent to:
    # chorogrid.plot(county_data.fips, county_data["Median_value_of_owner-occupied_housing_units_2009-2013"], ...)
    x="fips", y="Median_value_of_owner-occupied_housing_units_2009-2013",
    quantile=True)
```
